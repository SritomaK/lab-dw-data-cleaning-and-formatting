## Exercise 2: Cleaning invalid Values

The dataset contains columns with inconsistent and incorrect values that could affect the accuracy of our analysis. Therefore, we need to clean these columns to ensure that they only contain valid data.

Note that this exercise will focus only on cleaning inconsistent values and will not involve handling null values (NaN or None).

*Hint*:
- *Gender column contains various inconsistent values such as "F", "M", "Femal", "Male", "female", which need to be standardized, for example, to "M" and "F".*
- *State abbreviations be can replaced with its full name, for example "AZ": "Arizona", "Cali": "California", "WA": "Washington"*
- *In education, "Bachelors" could be replaced by "Bachelor"*
- *In Customer Lifetime Value, delete the `%` character*
- *In vehicle class, "Sports Car", "Luxury SUV" and "Luxury Car" could be replaced by "Luxury"*

```python
# Gender column contains various inconsistent values such as "F", "M", "Femal", "Male", "female", which need to be standardized, for example, to "M" and "F".

# data['gender'] = data['gender'].astype('str') 
# data['gender'] = list(map(lambda x: x.upper(), data['gender']))
# #list(map(len, data.columns))

# # Now define a function to clean the column
# def clean(x):
#     if x in ['M', 'MALE']:
#         return 'M'
#     elif x.startswith('F'):
#         return 'F'
#     else:
#         return x

# data['gender'] = list(map(clean, data['gender']))
```

```python
def clean(x):
    if x in ['M', 'Male']:
        return 'M'
    elif x in ['F', 'Femal', 'female']:
        return 'F'
    else:
        return x

data['gender'] = list(map(clean, data['gender']))
```

```python
data['gender'].isna().sum()
```

```python
data
```

```python
data['gender'].value_counts()
```

```python
data['state'].unique()
```

```python
#State abbreviations be can replaced with its full name, for example "AZ": "Arizona", "Cali": "California", "WA": "Washington"

#data['state'] = data['state'].astype('str') 

# Now define a function to clean the column
def clean_State(x):
    if x=='AZ':
        return 'Arizona'
    elif x=='Cali':
        return 'California'
    elif x=='WA':
        return 'Washington'
    else:
        return x

data['state'] = list(map(clean_State, data['state']))
```

```python
data['state'].unique()
```

```python
data['state'].value_counts()
```

```python
data['state'].isna().sum()
```

```python
#In education, "Bachelors" could be replaced by "Bachelor"

data['education'] = data['education'].str.replace('Bachelors','Bachelor')
data['education'].value_counts()
```

```python
#In Customer Lifetime Value, delete the % character

character_to_remove = '%'

# Useing the str.replace method to remove the character
data['customer_lifetime_value'] = data['customer_lifetime_value'].str.replace(character_to_remove, '')
```

```python
data
```

```python
#dividing the values in customer_lifetime_value by 100 because they are percentages
data['customer_lifetime_value'] = pd.to_numeric(data['customer_lifetime_value'], errors='coerce')
data['customer_lifetime_value'] = data['customer_lifetime_value'] / 100
data
```

```python
#in vehicle class, "Sports Car", "Luxury SUV" and "Luxury Car" could be replaced by "Luxury"

replacement_dict = {'Sports Car': 'Luxury','Luxury SUV': 'Luxury','Luxury Car': 'Luxury'}

data['vehicle_class'] = data['vehicle_class'].replace(replacement_dict)
```

```python
data['vehicle_class'].unique()
```
